```python
from autoscraper import AutoScraper
```

```python
url = 'https://www.amazon.in/s?k=watch'

wanted_list = ["Vintage Digital Grey Dial Unisex A-158WA-1Q ( D011 )",'₹1,695',"38,901"]
```

```python
scraper = AutoScraper()
result = scraper.build(url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_lxgs':"Title","rule_5159":'Price'})
scraper.keep_rules(['rule_lxgs','rule_5159'])
scraper.save('amazon-search')
```

```python
res = scraper.get_result_similar("https://www.amazon.in/s?k=casio+watches+for+men",group_by_alias=True)
```

# LinkedIN Scraping

```python
from selenium import webdriver
import time
import pandas as pd
import os

from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup

# import json
```

```python
url = "https://www.linkedin.com/jobs/search/?currentJobId=3830357434&f_E=2&keywords=data%20analyst&origin=JOB_SEARCH_PAGE_JOB_FILTER&refresh=true"
```

```python
path = r"C:/Program Files/Google/Chrome/Application/chrome.exe"
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--no-sandbox')
options.add_argument('--disable-dve-shm-usage')

options.binary_location = path

driver = webdriver.Chrome(options=options)

driver.get(url)
driver.implicitly_wait(10)
```

```python
n = driver.find_elements(By.CLASS_NAME,"results-context-header__job-count")[0].text
```

```python
n
```

```python
# n = n.replace('+', '').replace(',', '')
y = pd.to_numeric(n)
y
```

```python
# for infinite scroll function
i = 2
while i<=16:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    i = i+1
    
    try:
        x = driver.find_elements(By.XPATH,"button[@aria-label='Load more result']")
        driver.execute_script("arguments[0].click();")
        time.sleep(3)
    except:
        pass
    time.sleep(4)
    
```

```python
company_name=[]
```

```python
for j in range(y):
    try:
        company = driver.find_elements(By.CLASS_NAME,"base-search-card__subtitle")[j].text
        company_name.append(company)
    
    except IndexError:
        print("done")
```

```python
company_name
```

```python
len(company_name)
```

```python
title_name = []
```

```python
for j in range(y):
    try:
        title = driver.find_elements(By.CLASS_NAME,"base-search-card__title")[j].text
        title_name.append(title)
    
    except IndexError:
        print("done")
```

```python
title_name
```

```python
len(title_name)
```

```python
com_list = pd.DataFrame(company_name,columns=['company'])
title_list = pd.DataFrame(title_name,columns=['title'])
```

```python
job_list = com_list.join(title_list)
job_list
```

```python
job_list.head()
```

```python
job_list = job_list.to_csv("linkedin_jobs.csv")
```
